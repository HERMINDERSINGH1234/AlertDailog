```
# kernal density estimation plot ...
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```
# create dataset
dataset=np.random.randn(25)
# create the another rugplot
sns.rugplot(dataset);
# set-up the x-axis for the plot
x_min = dataset.min()-2
x_max = dataset.max()+2

# 100 equally spaced point from the x_min to the x_max
x_axis= np.linspace(x_min,x_max,100)

url= 'https://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'
bandwidth= ((4*dataset.std()**5)/(3*len(dataset)))**.2

# create an empty kernal list
kernal_list=[]
# plot the each basis function ..
for data_point in dataset:
  # create the kernal for the each point and append to list..
  kernal = stats.norm(data_point,bandwidth).pdf(x_axis)
  kernal_list.append(kernal)

  # scale the plotting
  kernal= kernal/kernal.max()
  kernal=kernal*.4
  plt.plot(x_axis,kernal,color='grey',alpha=0.5)
plt.ylim(0,1)
```
